```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

## **1) Load a gray level image with python statement.**

```
img = mpimg.imread('gray.jpg')
#gray = rgb2gray(img)
gray = img
```

## **2) Display it.**

```

#print(gray.shape)
plt.imshow(gray)
plt.show()

#plt.savefig('image_gray.jpeg')
```

## **3) Using only for loops**

# **a) flip the image in x axis.**


```
numrows = len(gray)   
numcols = len(gray[0])
fliped_image_x = np.zeros((numrows, numcols, 3)).astype('uint8')
fliped_image_y = np.zeros((numrows, numcols, 3)).astype('uint8')

for i in range(numrows):
  for j in range(numcols):
    fliped_image_x[i, j, 0] = gray[numrows - i -1, j, 0]
    fliped_image_x[i, j, 1] = fliped_image_x[i, j, 0]
    fliped_image_x[i, j, 2] = fliped_image_x[i, j, 0]
    
    
```

# **b) flip the image in y axis.**

```
for i in range(numrows):
  for j in range(numcols):
    fliped_image_y[i, j, 0] = fliped_image_x[i, numcols -  j - 1, 0]
    fliped_image_y[i, j, 1] = fliped_image_y[i, j, 0]
    fliped_image_y[i, j, 2] = fliped_image_y[i, j, 0]
```

# **c) Display the flipped images**

```
fig=plt.figure(figsize=(20, 10))

fig.add_subplot(1,3, 1)
plt.imshow(gray)
plt.title('Original Image', fontsize=12)



fig.add_subplot(1,3, 2)
plt.imshow(fliped_image_x)
plt.title('Flip the image in x axis.', fontsize=12)


fig.add_subplot(1,3, 3)
plt.imshow(fliped_image_y)
plt.title('Flip the image in y axis.', fontsize=12)

plt.show()
```

## **4) Save the image into anew file.**

```
plt.imsave('result.png',fliped_image_y)
fig.savefig('flip_result.png')
```

## **5) Using only for loops and no library functions, calculate and display following outputs.**

# **a) For all x,y in I, A(x,y) = 255-I(x,y)**

```
img = mpimg.imread('image.jpeg')
numrows = len(img)   
numcols = len(img[0])
A = np.zeros((numrows, numcols, 3)).astype('uint8')

for i in range(numrows):
  for j in range(numcols):
    A[i, j, 0] = 255 - img[i, j, 0]
    A[i, j, 1] = 255 - img[i, j, 1]
    A[i, j, 2] = 255 - img[i, j, 2]
```

# **b) For all x,y in I, B(x,y) = I(x,y)*2**

```
B = np.zeros((numrows, numcols, 3)).astype('uint8')

for i in range(numrows):
  for j in range(numcols):
    B[i, j, 0] = (img[i, j, 0]*2)
    B[i, j, 1] = (img[i, j, 1]*2)
    B[i, j, 2] = (img[i, j, 2]*2)

```

# **c) For all x,y in I, C(x,y)= I(x,y)-I(x-1,y-1)**

```
C = np.zeros((numrows, numcols, 3)).astype('uint8')
for i in range(1, numrows, 1):
  for j in range(1, numcols, 1):
      C[i - 1, j, 0] = img[i, j, 0] - img[i - 1, j - 1, 0]
      C[i - 1, j, 1] = img[i, j, 1] - img[i - 1, j - 1, 1]
      C[i - 1, j, 2] = img[i, j, 2] - img[i - 1, j - 1, 2]
```

```
fig=plt.figure(figsize=(20, 10))

fig.add_subplot(1,3, 1)
plt.imshow(A)
plt.title('A(x,y) = 255-I(x,y)', fontsize=12)



fig.add_subplot(1,3, 2)
plt.imshow(B)
plt.title('B(x,y) = I(x,y)*2', fontsize=12)


fig.add_subplot(1,3, 3)
plt.imshow(C)
plt.title('C(x,y)= I(x,y)-I(x-1,y-1)', fontsize=12)

plt.show()
fig.savefig('change_result.png')
```
